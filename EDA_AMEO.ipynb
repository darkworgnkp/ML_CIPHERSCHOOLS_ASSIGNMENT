# Aspiring Mind Employment Outcome 2015 (AMEO)

The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate. Below mentioned table contains the details for the original dataset.

# Data Cleaning

```python
# drop unnecessary cols
ameo.drop(columns=['CollegeID','CollegeCityID','12graduation','GraduationYear','JobCity','CollegeCityTier','DOJ', 'DOL'],axis=1,inplace=True)
```

```python
ameo.columns 
```

```python
ameo.set_index('ID',inplace=True)
```

```python
ameo.head(50)
```

```python
# counting null or  -1 or 0 values
ameo.isin([0,-1,np.nan]).sum()
```

```python
ameo.replace(to_replace=0,value=np.nan,inplace=True)
# replacing 0 with nan(object cols) used to further remove rows with null values
ameo.dropna(subset=['10board','12board'],inplace=True)
```

```python
ameo.isin([0,-1,np.nan]).sum()
```

```python
ameo.replace(to_replace=-1,value = 0,inplace=True)
# replcaing null value(-1) with 0(numerical cols)
```

```python
# creating new columns with avg score and avg academic performance
ameo['avg_score']= (ameo['English']+ameo['Logical']+ameo['Quant'])/3
ameo['acadperf'] = (ameo['10percentage']+ameo['12percentage']+ameo['collegeGPA'])/3
```

# Univariate Analysis

```python
plt.figure(figsize=(16,8))  # pd over salary(target)
plt.title('Salary dist')
sns.distplot(numerical_ameo['Salary'])
plt.show()
```

```python
px.box(x='Gender',y='Salary',data_frame=ameo) #boxplot for salary lvl by gender (refer to image attached  or use nbviewer)
```

### Observation

- Salary >10,00,000 is very rare, especially in the first job, So these are considered as outliers and also it can be seen by the difference between mean and median.
- most of the person have avg sal around 300 k
- higher outliers are observed in males indicates that salaries above normal are earned more by males then females.
- median salary of males is slightly greater than that of females.

```python
plt.figure(figsize=(15,8))  # hist plot on 10 and 12 percentage
plt.subplot(1,2,1)
plt.title('Plot 1')
sns.distplot(numerical_ameo['10percentage'],color='r')
plt.subplot(1,2,2)
plt.title('Plot 2')
sns.distplot(numerical_ameo['12percentage'])
plt.show()
```

### Observation

- more no. of person have 85+ % in class 10 and 68+% in class 12.
- Most of the People belong to avg and above avg.
- both 10 and 12 percentage are normally distributed can be seen from both plot1,2.

```python
plt.figure(figsize=(7,5)) # hist plot on college tiers(1,2)
sns.countplot(numerical_ameo['CollegeTier'])
```

### Observation

- large no. of persons 3500+ are qualified from Tier 2 and less from Tier 1 colleges

```python
plt.figure(figsize=(15,5)) # hist plot on gender aassociated with both tiers
plt.subplot(1,3,1)
plt.title('Plot 1')
sns.countplot(categorical_ameo['Gender'])
t1 = ameo.Gender.where(ameo.CollegeTier == 1)
t2 = ameo.Gender.where(ameo.CollegeTier == 2)
plt.subplot(1,3,2)
plt.title('Plot 2')
sns.countplot(t1)
plt.xlabel('Gender Tier 1')
plt.subplot(1,3,3)
plt.title('Plot 3')
sns.countplot(t2)
plt.xlabel('Gender Tier 2')
```

### Observation

- No of Males are more then Females from 1 plot
- most of the males and females are from Tier 2 college 2,3 plots

# Bivariate Analysis

```python
plt.figure(figsize=(15,6)) # scatter plot over salary with college tier and gpa.
plt.subplot(1,2,1)
plt.title('Plot 1')
sns.scatterplot(data=ameo,x='collegeGPA',y='Salary',hue='Gender')
plt.subplot(1,2,2)
plt.title('Plot 2')
sns.scatterplot(data=ameo,x='CollegeTier',y='Salary',hue='Gender')
```

### Observation

- some of the males have high salary of 4000000 having gpa above 60 belong to both Tier 1 and 2 colleges
- Tier 1 College have higher avg salary then Tier 2 but there are more no of student that have high package in tier 2 colleges than 1
- person who are having clgGPA <20 were not able to get higher package i.e above 10lakh
- from plot 1 there are some male and females that have low gpa but are earning good (outlier)

```python
# scatter and regg plot over salary and avg_score(eng,quant,logical)
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.title('Plot 1')
plt.ylim(0,1000000)
plt.xlabel('Average score')
sns.scatterplot(x='avg_score',y='Salary',data=numerical_ameo)
plt.subplot(1,2,2)
plt.title('Plot 2')
plt.ylim(0,1000000)
plt.xlabel('Average score')
sns.regplot(x='avg_score',y='Salary',data=numerical_ameo)
```

### Observation

- linear positive correlation between salary and average(logical+quant+english) score is indicated

```python
plt.figure(figsize=(16,75)) # scatter plot over designation  and salary
plt.title('Salary & Designation')
sns.scatterplot(x='Salary',y='Designation',data=ameo,hue='Gender')
```
